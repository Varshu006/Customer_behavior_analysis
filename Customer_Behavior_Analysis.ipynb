```python
import sqlite3
import pandas as pd

# Ye function aapka baar-baar upar-niche ka kaam khud karega
def run_sql(query):
    conn = sqlite3.connect('customer.db')
    try:
        df_res = pd.read_sql_query(query, conn)
        conn.close()
        return df_res
    except Exception as e:
        conn.close()
        print(f"Error: {e}")
```

```python
run_sql("""
SELECT "Customer ID", "Previous Purchases"
FROM customer_data
WHERE "Discount Applied" = 'Yes'
  AND "Previous Purchases" >= (SELECT AVG("Previous Purchases") FROM customer_data);
""")
```

```python
run_sql("""
select "Item Purchased", AVG("Review Rating") as avg_rating
from customer_data
group by "Item Purchased"
order by avg_rating desc
limit 5 ;"""
)
```

```python
run_sql("""
SELECT "Shipping Type", AVG("Previous Purchases")
FROM customer_data
WHERE "Shipping Type" IN ('Standard', 'Express', 'standard', 'express')
GROUP BY "Shipping Type";
""")
```

```python
run_sql("""
select "Subscription Status", COUNT("Customer Id")as "Total Customers",
AVG("Previous Purchases") as "Average Purchases",
SUM("Previous Purchases") as "Total Purchases"
from customer_data
group by "Subscription Status"
order by "Total Purchases" , "Average Purchases" desc;
""")
```

```python
run_sql("""
WITH "customer Type" AS (
    SELECT "Customer ID",
           CASE
               WHEN "Previous Purchases" = 1 THEN 'New'
               WHEN "Previous Purchases" BETWEEN 2 AND 10 THEN 'Returning'
               ELSE 'Loyal'
           END AS "customer_segment"
    FROM customer_data
)
SELECT "customer_segment", COUNT(*) AS "number of customer"
FROM "customer Type"
GROUP BY "customer_segment"
ORDER BY "number of customer" DESC;
""")
```

```python
run_sql("""
WITH "Item Counts" AS (
    SELECT "Category",
           "Item Purchased",
           COUNT("Customer ID") AS "Total Orders",
           ROW_NUMBER() OVER(
               PARTITION BY "Category"
               ORDER BY COUNT("Customer ID") DESC
           ) AS "Item Rank"
    FROM customer_data
    GROUP BY "Category", "Item Purchased"
)
SELECT "Item Rank", "Category", "Item Purchased", "Total Orders"
FROM "Item Counts";
""")
```

```python
run_sql("""
select "Subscroption Status",
COUNT("Customer ID")as "Total Customers"
from customer_data
where "Previos Purchases" > 5
group by "Subscription Status"
""")
```

```python
run_sql("""
SELECT
    CASE
        WHEN Age BETWEEN 18 AND 30 THEN 'Young'
        WHEN Age BETWEEN 31 AND 50 THEN 'Adult'
        ELSE 'Senior'
    END AS "Age Group",
    SUM("Previous Purchases") AS "Total Revenue"
FROM customer_data
GROUP BY "Age Group"
ORDER BY "Total Revenue" DESC;
""")
```
